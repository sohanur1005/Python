```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)
print(df)
```

```python
column = df['Age']
print(column)
```

```python
row = df.loc[1]  # Selects the second row (index 1)
print(row) 
```

```python
subset = df.loc[1:3, ['Name', 'City']]  # Select rows 1 to 3 and columns 'Name' and 'City'
print(subset)
```

```python
df.isnull()
```

```python
df.fillna(0, inplace=True)
df
```

```python
df.dropna(inplace=True)
```

```python
grouped = df.groupby('City').agg({'Age': ['mean']})
print(grouped)
```

```python
grouped = df.groupby('City').agg({'Age': ['mean', 'min', 'max']})
print(grouped)
```
